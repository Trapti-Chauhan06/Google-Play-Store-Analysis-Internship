## Convert object to Numeric

```python
apps_df['Reviews']=pd.to_numeric(apps_df['Reviews'])
```

```python
apps_df['Reviews']=pd.to_numeric(apps_df['Reviews'])
apps_df['Reviews'].dtype
```

```python
apps_df['Installs']=(apps_df['Installs'].str.replace(',',' ',regex=False).str.replace('+','',regex=False))
apps_df['Installs'] = apps_df['Installs'].str.replace(' ', '', regex=False)
apps_df['Installs'] = pd.to_numeric(apps_df['Installs'])
apps_df['Installs'].head()
```

```python
apps_df['Price'].unique()
```

```python
apps_df['Price'].sample(20)
```

```python
apps_df['Price'] = (
    apps_df['Price']
    .astype(str)
    .str.replace('$', '', regex=False)
)

apps_df['Price'] = pd.to_numeric(apps_df['Price'], errors='coerce')
```

```python
apps_df['Price'].dtype
apps_df['Price'].head(10)
```

```python
apps_df['Price'].isnull().sum()
```

```python
apps_df=apps_df[apps_df['Size']!='Varies with devide']
```

```python
print(apps_df['Size'].dtype)
```

```python
def convert_size(size):
    if pd.isna(size):
        return None
    size = str(size)

    if size == 'Varies with device':
        return None
    elif size.endswith('M'):
        return float(size[:-1])
    elif size.endswith('k'):
        return float(size[:-1]) / 1024
    else:
        return None

apps_df['Size'] = apps_df['Size'].apply(convert_size)


```

```python
apps_df['Size'].head()
```

```python
apps_df['Last Updated']=pd.to_datetime(apps_df['Last Updated'])
```

```python
apps_df['Month']=apps_df['Last Updated'].dt.month_name()
```

```python
apps_df[['Last Updated','Month']].head()
```

```python
apps_df["Android Ver"] = apps_df["Android Ver"].replace("Varies with device", "0.0")
apps_df["Android Ver"] = apps_df["Android Ver"].str.extract(r'(\d+\.?\d*)')
apps_df["Android Ver"] = pd.to_numeric(apps_df["Android Ver"], errors="coerce")
apps_df['Android Ver']
```

## Filter the Data

```python
filtered_df=apps_df[(apps_df['Rating']>=4.0) & (apps_df['Size']>=10) & (apps_df['Month'] == 'January')]
```

```python
filtered_df.shape
```

## Find Top 10 Categories by Installs

```python
top_categories=(filtered_df.groupby('Category')['Installs']
.sum()
.sort_values(ascending=False)
.head(10)
.index)
```

## Calculate Required Values

```python
chart_data=(filtered_df[filtered_df['Category'].isin(top_categories)].groupby('Category').agg(Average_rating=('Rating','mean'),
Total_Reviews=('Reviews','sum'))
.reset_index())
```

```python
chart_data
```

## Display Chart Only Between 3 PM and 5 PM IST

```python
from datetime import datetime
import pytz
ist =pytz.timezone("asia/Kolkata")
current_time= datetime.now(ist)
if 15<=current_time.hour<17:
    print("Graph can be displayed.:")
else:
    print("Graph can only be displayed between 3 PM and 5 PM IST.")
    plt.show()
    
```

## Create the Grouped Bar Chart

```python
x =np.arange(len(chart_data['Category']))
width=0.4
fig, ax1 =plt.subplots(figsize=(14,6))
ax1.bar(x - width/2, chart_data['Average_rating'],width,label='Average_rating')
ax1.set_ylabel('Average rating')
ax1.set_ylim(0,5)
ax2 =ax1.twinx()
ax2.bar(x + width/2, chart_data['Total_Reviews'], width, label='Total Reviews')
ax2.set_ylabel('Total Reviews')
ax1.set_xticks(x)
ax1.set_xticklabels(chart_data['Category'], rotation=45)
plt.title('Average rating vs Total Reviews (Top 10 categories)') 
fig.legend(loc = 'upper right')
plt.tight_layout()



```
